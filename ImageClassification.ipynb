```python

# Preprocessing
#1. resize 
#2. Flattening
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
#How do I convert matrix to vector? Flatten method
#flatten out pixels into one dimension
target = []
dataset = []
flat_data = []
DATADIR = '/content/dataset'
CATEGORIES= ['african wild dogs','buffalo','cheetah in trail', 'elephant', 'lion']


for category in CATEGORIES:
  class_num= CATEGORIES.index(category) #Label Encoding the values
  path = os.path.join(DATADIR, category) #create path to use all the images
  print(path)
  for img in os.listdir(path):
    imread(os.path.join(path,img))
    img_array = imread(os.path.join(path,img))
    #print (img_array)
    #plt.imshow(img_array)
    img_resized = resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    dataset.append(img_resized)
    target.append(class_num)

flat_data = np.array(flat_data)
target = np.array(target)
dataset = np.array(dataset)
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(flat_data,target, 
                                                   test_size=0.3, random_state= 30)
```

```python
#build model
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel': ['linear']}, 
              {'C':[1,10,100,1000],'gamma':[0.001,0.001], 'kernel':['rbf']},
]
svc = svm.SVC(probability = True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
#use classification algorithm
```
